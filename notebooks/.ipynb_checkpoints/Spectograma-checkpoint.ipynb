```python
import os, sys, time
import matplotlib.pyplot as plt
from scipy.io import wavfile
from scipy import signal
from tqdm import tqdm
import numpy as np
%matplotlib inline
```

```python
def get_file_paths(root_dir, return_user_id = False):
    ''' 
    Obtem o nome dos arquivos dentro uma sequencia de diretoriso na forma:
    [ROOT_DIR]/[USER_ID]/[INTERVIEW_ID]/[AUDIO_CUT_NAME].wav
    '''
    
    ids, file_path = [], []
    
    for id in os.listdir(root_dir):
        id_path = os.path.join(root_dir, id)
        for rec in os.listdir(id_path):
            cut_path = os.path.join(id_path, rec)
            for cut in os.listdir(cut_path):
                ids.append(id)
                file_path.append(os.path.join(cut_path, cut))
    if return_user_id:
        return ids, file_path
    else:
        return file_path
```

```python
class spec_transform(object):
    def __init__(self, wav_bits=16, t_length = 8.5):
        self.wav_bits = wav_bits
        self.t_length = t_length
    
    def __str__(self):
        print('spec_transform: %d bits, %d secs' % (self.wav_bits, self.t_length))
    
    
    def read_wav(self, wav_fname):
        srate, wav = wavfile.read(wav_fname)
        return srate, wav
    
    def normalize_wav(self, wav):
        return wav / (2 ** (self.wav_bits - 1))
    
    def pad_wav(self, wav, srate):
        nb_pad_frames = int(self.t_length * srate - wav.shape[0])
        return np.concatenate((wav, np.zeros(nb_pad_frames)))

    
    def pspectogram(self, wav, srate, window_size, step_size):
        nperseg = int(round(window_size * srate / 1e3))
        noverlap = int(round(step_size * srate / 1e3))

        f, t, spec = signal.spectrogram(wav, fs=srate, window='hann', nperseg=nperseg,  
                                        noverlap=noverlap, detrend=False)

        return f, t, spec

    def clip_spec(f, t, spec, log_freq=True, log_spec=True, min_freq=4, eps=1e-15):
        if log_freq:
            f = np.log(f + eps)
        if log_spec:
            spec = np.log(spec + eps)

        spec = spec[f >= min_freq, :]
        f = f[f >= min_freq]

        return f, t, spec
    
    def transform(self, wav_fname, window_size=64, step_size=32, min_freq=4, eps=1e-15):
        srate, wav = self.read_wav(wav_fname)
        normalized_wav = self.normalize_wav(wav)
        padded_wav = self.pad_wav(normalized_wav, srate)
        f, t, spec = self.pspectogram(padded_wav, srate, window_size=window_size, step_size=step_size)
        f, t, spec = self.clip_spec(f, t, spec, min_freq=min_freq, eps=eps)
        return np.array(spec)
```
