```python
import os, sys, time
import matplotlib.pyplot as plt
import librosa
from tqdm import tqdm
import numpy as np
from librosa import display

def get_mel_spec(wavf, length_out = 10):
    wav, sr = librosa.load(wavf)
    wav = librosa.util.fix_length(wav, int(sr*length_out))
    return librosa.power_to_db(librosa.feature.melspectrogram(wav, sr = sr))
    
def save_mel(fname, S):
    fig,ax = plt.subplots(1, figsize=(2.8828125*1.785, 1*1.78))
    fig.subplots_adjust(left=0,right=1,bottom=0,top=1)
    display.specshow(S, fmax=8000, cmap='gray')
    ax.axis('off')
    fig.savefig(fname, frameon=False)
```

```python
save_mel('bla.png', get_mel_spec('../dados/wav/id10001/1zcIwhmdeo4/00003.wav'))
```
